```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Netflix Userbase.csv")

df.head()
```

```python
print("Shape of Dataset:", df.shape)

print("\nColumns:")
print(df.columns)

print("\nInformation:")
df.info()

print("\nMissing Values:")
print(df.isnull().sum())

print("\nStatistical Summary:")
display(df.describe(include="all"))
```

```python
# Display first 10 rows

df.head(10)
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Subscription Type", data=df)
plt.title("Subscription Type Distribution")
plt.xticks(rotation=20)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Gender", data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Device", data=df)
plt.title("Device Distribution")
plt.xticks(rotation=20)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], bins=10, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="Subscription Type", y="Monthly Revenue", data=df)
plt.title("Monthly Revenue by Subscription")
plt.xticks(rotation=20)
plt.show()
```

```python
plt.figure(figsize=(10,5))
df["Country"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Countries by Users")
plt.ylabel("Number of Users")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=["int64","float64"])

plt.figure(figsize=(6,4))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
